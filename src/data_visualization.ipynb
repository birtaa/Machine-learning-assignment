```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Read data
df = pd.read_csv(r"D:\gitClones\Machine-learning-assignment\data\WineQT_missing.csv")

# Set up the main figure for quality distributions
plt.figure(figsize=(15, 10))

# 1. Quality Distribution
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='quality', bins=len(df['quality'].unique()))
plt.title('Wine Quality Distribution')
plt.xlabel('Quality Score')
plt.ylabel('Count')

# 2. Alcohol vs Quality
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='quality', y='alcohol')
plt.title('Alcohol Content by Quality')
plt.xlabel('Quality Score')
plt.ylabel('Alcohol %')

# 3. Fixed Acidity Distribution by Quality
plt.subplot(2, 2, 3)
sns.violinplot(data=df, x='quality', y='fixed acidity')
plt.title('Fixed Acidity by Quality')
plt.xlabel('Quality Score')
plt.ylabel('Fixed Acidity')

# 4. Sulfur Dioxide Distribution
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='quality', y='total sulfur dioxide')
plt.title('Sulfur Dioxide by Quality')
plt.xlabel('Quality Score')
plt.ylabel('Total Sulfur Dioxide')

plt.tight_layout()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 8))
correlation = df.corr()
mask = np.triu(np.ones_like(correlation, dtype=bool))
sns.heatmap(correlation, mask=mask, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()

# Pairplot for key features
features_of_interest = ['quality', 'alcohol', 'fixed acidity', 'volatile acidity', 'citric acid']
sns.pairplot(df[features_of_interest], hue='quality', diag_kind='hist')
plt.show()
```

```python


# Srekli deikenler listesi
continuous_vars = ['fixed acidity', 'volatile acidity', 'citric acid', 
                  'residual sugar', 'chlorides', 'free sulfur dioxide',
                  'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

# Plot ayarlar
plt.figure(figsize=(15, 30))

# Her srekli deiken iin line plot
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(len(continuous_vars), 1, i)
    
    # Eksik deerleri filtrele ve srala
    # plot_data = df[var].dropna().sort_values()
    plot_data = df[var].dropna()
    
    plt.plot(range(len(plot_data)), plot_data)
    plt.title(f'Distribution of {var}')
    plt.xlabel('Observation Number')
    plt.ylabel(var)
    plt.grid(True)

plt.tight_layout()
plt.show()

# statistiksel zet
print("\nstatistiksel zet:")
print(df[continuous_vars].describe())
```
